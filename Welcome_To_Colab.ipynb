```
import figures as f
print("1 cube 2 cuboid 3 cylinder 4 cone 5 sphere")
c=int(input("Enter figure:"))

print("1 CSA 2 TSA 3 Volume")
o=int(input("Enter operation:"))

if c==1:
  a=float(input("Side:"))
  r=f.cube(a)

elif c==2:
  l,b,h=map(float,input("l b h:").split())
  r=f.cuboid(l,b,h)

elif c==3:
  r1,h=map(float,input("r h:").split())
  r=f.cylinder(r1,h)

elif c==4:
  r1,l,h=map(float,input("r l h:").split())
  r=f.cone(r1,l,h)

elif c==5:
  r1=float(input("radius:"))
  r=f.sphere(r1)

print("result=",r[0-1])
```
