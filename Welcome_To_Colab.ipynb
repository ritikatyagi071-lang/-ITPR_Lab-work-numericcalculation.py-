```
import math
def cube(a): return 4*a*a,6*a*a,a**3
def cuboid(l,b,h): return 2*h*(l+b),2*(l*b+b*h+l*h),l*b*h
def cylinder(r,h): return 2*math.pi*r*h,2*math.pi*r*(r+h),math.pi*r*r*h
def cone(r,l,h): return math.pi*r*l,math.pi*r*(r+1),(1/3)*math.pi*r*r*h
def sphere(r): return 4*math.pi*r*r,4*math.pi*r*r,(4/3)*math.pi*r**3
```
